#  EM (Expectation-Maximization) algorithm applied to Gaussian Mixtures

```julia
using PyPlot
using Random
```

```julia
# number of data points
N = 1000
# number of dimensins
n = 2
# generate random data
Random.seed!(1234)

x = randn(n,N)
k = 3
xc_true = 3 * randn(n,k)

for i = 1:N
    ic = rand(1:k)
    x[:,i] = xc_true[:,ic] + randn(2)
end

plot(x[1,:],x[2,:],"k.",markersize = 1);
x
```


A Gaussian pdf is given by its mean $\mathbf \mu$ and its covariance matrix $\mathbf \Sigma$

$$
p(\mathbf x) = \frac{1}{\sqrt{2\pi \det(\mathbf \Sigma)}} \exp\left(-\frac{1}{2} (\mathbf x - \mathbf μ)^T  \mathbf Σ^{-1} (\mathbf x - \mathbf μ) \right)
$$

```julia
Gaussian(x,μ,Σ) =  1/(sqrt(2π * det(Σ))) * exp(- (x - μ)' * (Σ \ (x - μ) )/2 )
```

A gaussian mixture model is a linear combination of Gaussians

$$
p'(\mathbf x) = \sum_{l=1}^k \frac{\phi_l}{\sqrt{2\pi \det(\mathbf \Sigma_l)}} \exp\left(-\frac{1}{2} (\mathbf x - \mathbf μ_l)^T  \mathbf Σ_l^{-1} (\mathbf x - \mathbf μ_l) \right)
$$

where all coefficients $\phi_l$ are positive and their sum is 1.

```julia
function GMM(x,ϕ,μ,Σ)
    GMM = 0.
    for l = 1:size(μ,2)
        GMM += ϕ[l] * Gaussian(x,μ[:,l],Σ[:,:,l])
    end
    
    return GMM
end
```

An example

```julia
# initialize
n = 2
μ = 2*randn(n,k)
Σ = zeros(n,n,k);
Σ[1,1,:] .= 1 
Σ[2,2,:] .= 1
ϕ = ones(k)/k

gx = range(-10,stop=10,length=100);
gy = range(-10,stop=10,length=100);

p = zeros(length(gx),length(gy));
pk = zeros(length(gx),length(gy),k);

for j = 1:length(gy)
    for i = 1:length(gx)
        p[i,j] = GMM([gx[i]; gy[j]],ϕ,μ,Σ);
        for l = 1:k
            pk[i,j,l] = Gaussian([gx[i]; gy[j]],μ[:,l],Σ[:,:,l]); 
        end
    end
end


contourf(gx,gy,copy(p'));
```

```julia
function plotGaussian(μ,Σ,color)
    gx = range(-10,stop=10,length=100);
    gy = range(-10,stop=10,length=100);

    p = zeros(length(gx),length(gy));
    #@show μ
    for j = 1:length(gy)
        for i = 1:length(gx)
            p[i,j] = Gaussian([gx[i]; gy[j]],μ,Σ); 
        end
    end
    
    contourf(gx,gy,copy(p'),levels = [0.1,100],colors=color,alpha=0.2)
end
    
function plotGaussians(μ,Σ,colors = ["r","g","b","c"])
    for l = 1:size(μ,2)
        plotGaussian(μ[:,l],Σ[:,:,l],colors[l])
    end
end

plotGaussians(μ,Σ)
```

Sampling from a gaussian mixture
* choose a $z$, an integer number from 1 to $k$. The probability of choosing $z = j$ is $\phi_j$ ($1 \le j \le k$).
* sample from the Gaussian distribution with mean $\mathbf \mu_j$ and covariance $\mathbf \Sigma_j$ where $j = z$.


As the data is unlabeld, we do not known from which Gaussian a particular data point is sampled from: $z$ is thus unknown.


## Expectation step

For each sample $i$ and Gaussian $j$, compute the weights $w^{(i)}_j$ wich corresponds to the probability that the sample
$ \mathbf x^{(i)}$ comes from the Gaussian $j$

$$
w^{(i)}_j = P(z^{(i)} = j \mid \mathbf x^{(i)})
$$

The Bayes rule allows us to compute these weights:


$$\begin{eqnarray} 
P(z^{(i)} = j \mid \mathbf x^{(i)}) &=& \frac{P(\mathbf x^{(i)} \mid z^{(i)} = j) \; P(z^{(i)} = j) } {\sum_{l=1}^k P(\mathbf x^{(i)} \mid z^{(i)} = l) \; P(z^{(i)} = l) }  \\
&=&  \frac{P(\mathbf x^{(i)} \mid z^{(i)} = j) \; \phi_j } {\sum_{l=1}^k P(\mathbf x^{(i)} \mid z^{(i)} = l) \; \phi_l }
\end{eqnarray}$$

where

$$
P(\mathbf x^{(i)} \mid z^{(i)} = j) = 
\frac{1}{\sqrt{2\pi \det(\mathbf \Sigma_j)}}
\exp\left(-\frac{1}{2} (\mathbf x - \mathbf μ_j)^T  \mathbf Σ_j^{-1} (\mathbf x - \mathbf μ_j) \right)
$$


```julia
function expectation_step(x,ϕ,μ,Σ)
    k = length(ϕ)
    N = size(x,2)
    w = zeros(k,N)

    for i = 1:N
        for j = 1:k
            w[j,i] = Gaussian(x[:,i],μ[:,j],Σ[:,:,j]) * ϕ[j]
        end
    
        w[:,i] = w[:,i]/sum(w[:,i])
    end

    return w
end
```

```julia
w = expectation_step(x,ϕ,μ,Σ) 

for l = 1:k
    subplot(2,2,l)
    scatter(x[1,:],x[2,:],10,w[l,:]); colorbar();
    plotGaussian(μ[:,l],Σ[:,:,l],"b")
end
```

Weighted mean assuming the sum of all elements of `wn` is 1.

```julia
function wmean(x,wn)
    μ = zeros(size(x,1));
    for i = 1:size(x,2)
       μ += wn[i] * x[:,i]
    end
    return μ
end
function wcovar(x,wn,μ = wmean(x,wn))
    Σ = zeros(length(μ),length(μ));
    for i = 1:size(x,2)
       Σ += wn[i] * (x[:,i] - μ) * (x[:,i] - μ)'
    end
    return Σ
end
    
```

## Maximization step

$$
\phi_j = \frac{1}{N} \sum_{i=1}^N  w^{(i)}_j
$$

Weighted mean:

$$
\mathbf \mu_j =  \frac{1}{ \sum_{i=1}^N  w^{(i)}_j  }  \sum_{i=1}^N  w^{(i)}_j  \mathbf x^{(i)}
$$

Weighted covariance:

$$
\mathbf \Sigma_j =  \frac{1}{ \sum_{i=1}^N  w^{(i)}_j  }  \sum_{i=1}^N  w^{(i)}_j  (\mathbf x^{(i)} - \mathbf μ_j) (\mathbf x^{(i)} - \mathbf μ_j)^T
$$



```julia
function maximization_step!(x,w,ϕ,μ,Σ)
    k = length(ϕ)
    N = size(x,2)

    for j = 1:k
        sumw = sum(w[j,:])
        # normalized weights 
        wn = w[j,:]/sumw
        
        ϕ[j] = sumw/N        
        μ[:,j] = wmean(x,wn)
        Σ[:,:,j] = wcovar(x,wn,μ[:,j])
    end
end
```

```julia
k = 3
μ = 2*randn(n,k)
Σ = zeros(n,n,k);
Σ[1,1,:] .= 1
Σ[2,2,:] .= 1
ϕ = ones(k)/k

fig = figure()

for niter = 1:10
    w = expectation_step(x,ϕ,μ,Σ)  

    maximization_step!(x,w,ϕ,μ,Σ)
    
    if niter % 1 == 0
        clf()
        plot(x[1,:],x[2,:],"k.",markersize=0.5)
        plotGaussians(μ,Σ)
        sleep(1)
        title("niter = $(niter)")
        IJulia.clear_output(true)
        display(fig)
    end
end
```
