# Linear Discriminant Analysis



* Assume that we have a set of vectors $\mathbf x^{(i)}$ and for each vector we have a label $y_i$ which is either 0 or 1.

Means and variances for each class

for $y = 0$

$$
\mathbf \mu_0 = \frac{1}{n_0} \sum_{i \in C_0} \mathbf x^{(i)}
$$

$$
\Sigma_0 =  \frac{1}{n_0-1} \sum_{i \in C_0} (\mathbf x^{(i)} - \mathbf \mu_0)^T (\mathbf x^{(i)} - \mathbf \mu_0)
$$

and for  $y = 1$

$$
\mathbf \mu_1 = \frac{1}{n_1} \sum_{i \in C_1} \mathbf x^{(i)}
$$

$$
\Sigma_1 =  \frac{1}{n_1-1} \sum_{i \in C_1} (\mathbf x^{(i)} - \mathbf \mu_1)^T (\mathbf x^{(i)} - \mathbf \mu_1)
$$



$$
\mathbf w = \mathrm{argmax}_w \frac{ (\mathbf \mu_0^T \mathbf w - \mathbf \mu_1^T \mathbf w)^2 } 
{n_0 \mathbf w^T \mathbf \Sigma_0  \mathbf w + n_1 \mathbf w^T \mathbf \Sigma_1  \mathbf w}
$$

$$
= \frac{ ((\mathbf \mu_O  - \mathbf \mu_1)^T \mathbf w)^2 } 
{\mathbf w^T (n_0  \mathbf \Sigma_0  + n_1  \mathbf \Sigma_1)  \mathbf w}
$$

$$
= \frac{ (\mathbf m^T \mathbf w)^2 } 
{\mathbf w^T \mathbf S  \mathbf w}
$$

$\mathbf S = (n_0  \mathbf \Sigma_0  + n_1  \mathbf \Sigma_1) $ and $\mathbf m = \mathbf \mu_O  - \mathbf \mu_1$.
Just as $\Sigma_0$  and $\mathbf \Sigma_1$, $\mathbf S$ is a symetric positive defined matrix.

$$
= \frac{ (\mathbf m^T \mathbf w)^2 } 
{\mathbf w^T \mathbf R^T \mathbf R  \mathbf w}
$$

where $\mathbf S = \mathbf R^T \mathbf R$ (square root decomposition)


$$
= \frac{ (\mathbf m^T \mathbf R^{-1} \mathbf v)^2 } 
{\mathbf v^T {\mathbf R^T}^{-1} \mathbf R^T \mathbf R  \mathbf R^{-1} \mathbf v}
$$

where $\mathbf v = \mathbf R \mathbf w$ (transformed vector)

$$
= \frac{ (\mathbf m^T \mathbf R^{-1} \mathbf v)^2 } 
{\mathbf v^T  \mathbf v}
$$

$\mathbf v^T  \mathbf v = \| \mathbf v \|^2$

$$
= \left( \frac{ \mathbf m^T \mathbf R^{-1} \mathbf v} 
{\| \mathbf v \|} \right)^2
= 
\left( \frac{ ({\mathbf R^{-1}}^T \mathbf m)^T  \mathbf v} 
{\| \mathbf v \|} \right)^2
$$

The function to maximize could be rewritten as the project between two vectors. The projection is maximum if they are a multiple of each other.

$$
\mathbf v = a \; {\mathbf R^{-1}}^T \mathbf m = a; {\mathbf R^{-1}}^T (\mathbf \mu_0 - \mathbf \mu_1)
$$

where $a$ is a arbitrary scalar.

$$
\mathbf w = \mathbf R^{-1} \mathbf v =  a \; \mathbf R^{-1}  {\mathbf R^{-1}}^T (\mathbf \mu_0 - \mathbf \mu_1) = a \; \mathbf S^{-1} (\mathbf \mu_0 - \mathbf \mu_1)
$$

$$ 
\mathbf w = (n_0  \mathbf \Sigma_0  + n_1  \mathbf \Sigma_1)^{-1} (\mathbf \mu_0 - \mathbf \mu_1)
$$
